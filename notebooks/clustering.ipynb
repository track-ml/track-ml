## Step 2 - Fit Data to Clustering Model



```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, normalize
from collections import defaultdict
```

```python

temp_df = pd.DataFrame(normalize(df), columns = df.columns)
temp_df = StandardScaler().fit_transform(temp_df)


model = DBSCAN(min_samples=3,eps=0.7).fit(temp_df)
labels = model.labels_
labels

label_dict = dict(zip(df.index,labels))


res = defaultdict(list)
for key, val in sorted(label_dict.items()):
    res[val].append(key)

res
```
